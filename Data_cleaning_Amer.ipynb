```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

ufo = pd.read_csv('https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/ufo.csv')
ufo.head()
```

```python
ufo.info()
```

```python
ufo.isnull().sum()
```

```python
ufo.dropna().shape
```

```python
ufo['City'].fillna(value='CityNotMentioned',inplace=True)
```

```python
ufo['Colors Reported'].fillna(value='ColorsNotKnown',inplace=True)
ufo['Shape Reported'].fillna(value='UnknownShape',inplace=True)
```

```python
ufo.isnull().sum()
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/sample11.csv")
data
```

```python
mean_val_marks = data['Marks'].mean()
data['Marks'].fillna(value=mean_val_marks,inplace=True)
data
```

```python
mean_val_per = data['Percentage'].mean()
mean_val_per = round(mean_val_per,1)
data['Percentage'].fillna(value=mean_val_per,inplace=True)
data
```
