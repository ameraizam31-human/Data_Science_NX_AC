```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print(sns.__version__)
sns.get_dataset_names()
dataset = sns.load_dataset('iris')
dataset
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.title('Lineplot In Seaborn')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set_style('darkgrid',{'grid.color':'lightblue','grid.linestyle':'--'})   # -- dashed  :dotted  - solid
sns.lineplot(x='sepal_length', y='sepal_width', data=dataset)
plt.title('Lineplot in Seaborn', color='blue',fontsize=12)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.show()
```

```python
pokemon = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/pokemon.csv')
pokemon
```

```python
sns.scatterplot(x='Attack',y='Defense',data=pokemon)
plt.title('Scatterplot In Seaborn')
plt.show()
```

```python
sns.scatterplot(x='Attack',y='Defense',hue='Generation',data=pokemon)
plt.title('Scatterplot In Seaborn')
plt.show()
pokemon.Generation.value_counts().sort_index()
```

```python
colors = ['r','y','b','g','c','m']
sns.scatterplot(x='Attack',y='Defense',hue='Generation',palette=colors,data=pokemon)
plt.title('Scatterplot In Seaborn')
plt.show()
```

```python
colors = ['r','y','b','g','c','m']
sns.lmplot(x='Attack',y='Defense',hue='Generation',palette=colors,data=pokemon)
plt.title('Scatterplot In Seaborn')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Type 1',data=pokemon)
plt.title('Countplot In Seaborn')
plt.show()
```

```python
#HEATMAP - a colored table where colors represent data values
# helps to quickly identify hig, low, positive, or negative values

data = {
    'Hours_Studied':[2,3,4,5,6,7,8],
    'Marks_Scored':[40,50,60,70,80,90,100],
    'TV_Watched':[10,9,8,6,7,5,3]
}

df = pd.DataFrame(data)
df
```

```python
corr = df.corr()
corr
```

```python
sns.heatmap(corr,cmap="Blues",annot=True)  # 'viridis', 'coolwarm', 'Blues'
plt.title("Heatmap")
plt.show()
```

```python
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.show()
```
